```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]
myexplode = [0.2, 0, 0, 0]
plt.pie(y, labels = mylabels, colors = mycolors, explode = myexplode, shadow = True, startangle = 90)
plt.legend(title = "Four Fruits:")
plt.show() 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
Drowning_Accident = [1,12,25,31,56,65,79,123,546]
Ice_Cream_Sale = [20,40,60,80,100,120,140,160,180]
Drowning_Dict = {"Drowning_Accident": Drowning_Accident,
"Ice_Cream_Sale": Ice_Cream_Sale}
Drowning = pd.DataFrame(data=Drowning_Dict)
# Drowning
Drowning.plot(x="Ice_Cream_Sale", y="Drowning_Accident", kind="scatter")
plt.show()

correlation_beach = Drowning.corr()
print(correlation_beach)
```
