Day 6:
    
Measures of Dispersion ( Range, Variance [Error Measure], Standard Deviation)

    Variance and error measure are not the same, although they are related concepts.
    Variance is a measure of the variability or spread of a set of data points around the mean or expected value. It is calculated as the average of the squared differences of each data point from the mean. In statistics, variance is often used as a measure of the amount of information contained in a data set.
    On the other hand, an error measure is a metric that quantifies the difference between the observed values and the predicted values. It is commonly used in machine learning and statistical modeling to evaluate the accuracy of a model's predictions. Examples of error measures include mean squared error (MSE), root mean squared error (RMSE), mean absolute error (MAE), etc.
    While variance is a measure of the spread of the data, error measures are used to quantify the performance of a model in making predictions on new data. However, there is a relationship between the two concepts, as high variance in a model can lead to overfitting, which in turn can result in high prediction errors on new data.

```python
import numpy as np # pip install numpy
```

```python
data = np.array([3, 5, 6, 7, 9, 12])
data
```

```python
"""
l = sorted(data.tolist())
data_range = l[-1] - l[0] #MAXIMUM - MINIMUM values
"""
data_range = np.ptp(data) # calculate the range of the array

# print the range
print("Range =", data_range)
```

```python
'''
In statistics, when we use a sample to estimate the parameters of the population, we typically use the sample standard deviation and variance instead of the population standard deviation and variance. However, since the sample is only a subset of the population, the sample statistics tend to underestimate the population parameters.
To adjust for this bias, we use the Bessel's correction, which involves dividing the sum of squared deviations by n-1 instead of n. The parameter "ddof" stands for "Delta Degrees of Freedom" and determines the value to subtract from the sample size to obtain the degrees of freedom. When we set ddof=1, we are indicating that we want to use the Bessel's correction and that we have one less degree of freedom compared to the total number of samples.
In other words, when ddof=1, we are estimating the variance of a sample from a larger population, and we are using the sample variance as an unbiased estimator of the population variance. When ddof=0, we are computing the variance of a population from the data, and there is no need to correct for bias.
'''
```

```python
variance = np.var(data, ddof=1 ) #
print("Variance =",variance)
```

```python
std_deviation = np.std(data, ddof=1) #
print("Standard Deviation =",std_deviation)
```

Day 9:

    Predictive Time Series Forecasting


```python
import pandas as pd #ALREDY IMPORTED
import numpy as np #ALREDY IMPORTED
import matplotlib.pyplot as plt #ALREDY IMPORTED
from statsmodels.tsa.arima.model import ARIMA # pip install statsmodels
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf #ALREDY IMPORTED
```

```python
# Load the time series data
data = pd.read_csv('P_redict.csv', index_col='Month', parse_dates=True)
data
```

```python
# Plot the time series
plt.plot(data)
plt.title('My Time Series')
plt.show()
```

```python
# Take the first difference to make the time series stationary
diff = data.diff().dropna()
diff
```

```python
# Plot the differenced time series
plt.plot(diff)
plt.title('Differenced Time Series')
plt.show()
```

```python
# Plot the ACF and PACF to determine the orders of p and q
plot_acf(diff)
plot_pacf(diff)
plt.show()
```

```python
'''
Autocorrelation (ACF) and partial autocorrelation (PACF) are functions used to describe the correlation between a time series and its lags.

Autocorrelation is a measure of how a time series is correlated with itself at different lags. Specifically, it measures the correlation between a time series and its own values at previous time steps, for various lags. The autocorrelation function (ACF) is a plot of these correlations as a function of the lag. The ACF is a useful tool for identifying patterns and trends in the data, such as seasonal patterns or trends.

Partial autocorrelation, on the other hand, measures the correlation between a time series and its own values at previous time steps, but after removing the effects of shorter lags. This is useful for identifying the specific lag(s) that have a significant impact on the current value of the time series. The partial autocorrelation function (PACF) is a plot of these correlations as a function of the lag, after accounting for the effects of shorter lags.

In summary, both ACF and PACF are tools used to explore and understand the relationship between a time series and its own values at different lags. They are important for identifying the order of the ARIMA model to fit to the data, which in turn helps in forecasting the time series accurately.
'''
```

```python
# Fit the ARIMA model with orders (1,1,1)
model = ARIMA(data, order=(1,1,1))
result = model.fit()

# Print the summary of the model
print(result.summary())
```

```python
# Make predictions for the next 10 periods
forecast = result.forecast(steps=10)
forecast
```

```python
# Plot the forecasted values
plt.plot(data)
plt.plot(forecast, color='red')
plt.title('Forecasted Time Series')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller #ALREDY IMPORTED
```

```python
result = adfuller(data['Passengers']) # <-- Access the 'Passengers' column
result
```
